```python
import pandas as pd
import numpy as np 
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df['area'], df['price'], marker='*', color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']].values, df['price'])
```

```python
model.predict([[5400]])
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df['area'], df['price'], marker='*', color='blue')
plt.plot(df['area'], model.predict(df[['area']].values), color='red')
```

```python
# Enregistrement du model avec Pickle
import pickle
```

```python
# Écriture du model dans un fichier binaire
with open("model_pickle", "wb") as f:
    pickle.dump(model, f)
```

```python
# Ouverture du fichier et lecture du model
with open("model_pickle", "rb") as f:
    mp = pickle.load(f)
```

```python
mp.predict([[6000]])
```
